```python
import matplotlib.pyplot as plt
import seaborn as sns

# Using 'CF_attrition label' for clarity
plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")
colors = sns.color_palette('pastel')[0:2]

df['CF_attrition label'].value_counts().plot.pie(explode=[0.1,0], autopct='%1.1f%%', shadow=True, colors=colors)
plt.title('Overall Employee Attrition Percentage', fontsize=15)
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Department', hue='Attrition', palette='magma')
plt.title('Attrition Distribution across Departments', fontsize=15)
plt.xlabel('Department')
plt.ylabel('Count of Employees')
plt.legend(title='Attrition')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
# Ordering the age bands for a cleaner look
order = sorted(df['CF_age band'].unique())
sns.countplot(data=df, x='CF_age band', hue='Attrition', order=order, palette='viridis')
plt.title('Attrition by Age Band', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Attrition', y='Job Satisfaction', palette='Set2')
plt.title('Job Satisfaction Levels: Stayed vs Left', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Monthly Income', hue='Attrition', kde=True, element="step", palette='flare')
plt.title('Monthly Income Distribution by Attrition Status', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Over Time', hue='Attrition', palette='coolwarm')
plt.title('Impact of Overtime on Employee Attrition', fontsize=15)
plt.show()
```
